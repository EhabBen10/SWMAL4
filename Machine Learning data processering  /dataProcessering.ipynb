```python
import pandas as pd

# Define the file path
file_path = '../Data/heart.csv'

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path)
data = data[data.columns[0:12]]
data.head()
# Display the first few rows of the DataFrame
print(data.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Check unique values in 'Sex' column
print("Unique values in 'Sex' column before cleaning:", data['Sex'].unique())

# Check for missing values
missing_count = data['Sex'].isnull().sum()
print(f"Missing values in 'Sex': {missing_count}")

data_cleaned_Sex= data['Sex'].replace({'M': 0, 'F': 1})  # Encode 'M' -> 0, 'F' -> 1

# Compute statistics
labelStr = 'Mean: {0}\nMedian: {1}\nStd: {2}\nMin: {3}\nMax: {4}'.format(
    np.round(data_cleaned_Sex.mean(), 3),
    np.round(data_cleaned_Sex.median(), 3),
    np.round(data_cleaned_Sex.std(), 3),
    np.round(data_cleaned_Sex.min(), 3),
    np.round(data_cleaned_Sex.max(), 3)
)

# Plot
fig, ax = plt.subplots()
ax = data_cleaned_Sex.value_counts().sort_index().plot(kind='bar', rot=0, label=labelStr, color=['blue', 'red'])
ax.set_title("Sex Distribution")
ax.set_xlabel("Sex (0=Male, 1=Female)")
ax.set_ylabel("Frequency")
ax.set_xticks([0, 1])  # Ensure correct ticks
ax.set_xticklabels(['Male (0)', 'Female (1)'])
ax.legend(loc="upper right")

# Show the plot
plt.show()
```

```python
# Check unique values in 'Sex' column
print("Unique values in 'ExerciseAngina' column before cleaning:", data['ExerciseAngina'].unique())

# Check for missing values
missing_count = data['ExerciseAngina'].isnull().sum()
print(f"Missing values in 'ExerciseAngina': {missing_count}")

data_cleaned_ExerciseAngina= data['ExerciseAngina'].replace({'N': 0, 'Y': 1})  # Encode 'M' -> 0, 'F' -> 1

# Compute statistics
labelStr = 'Mean: {0}\nMedian: {1}\nStd: {2}\nMin: {3}\nMax: {4}'.format(
    np.round(data_cleaned_ExerciseAngina.mean(), 3),
    np.round(data_cleaned_ExerciseAngina.median(), 3),
    np.round(data_cleaned_ExerciseAngina.std(), 3),
    np.round(data_cleaned_ExerciseAngina.min(), 3),
    np.round(data_cleaned_ExerciseAngina.max(), 3)
)

# Plot
fig, ax = plt.subplots()
ax = data_cleaned_Sex.value_counts().sort_index().plot(kind='bar', rot=0, label=labelStr, color=['blue', 'red'])
ax.set_title("ExerciseAngina Distribution")
ax.set_xlabel("Sex (0=N, 1=Y)")
ax.set_ylabel("Frequency")
ax.set_xticks([0, 1])  # Ensure correct ticks
ax.set_xticklabels(['N (0)', 'Y (1)'])
ax.legend(loc="upper right")

# Show the plot
plt.show()
```

```python

ax = data.hist(bins=25, figsize=(15, 15))

# Iterate over each row of subplots
for row in ax:
    for subplot in row:
        # Get the column name from the subplot's title
        column = subplot.get_title()
        
        if column:  # Check if the subplot has a title
            # Calculate statistics directly from the dataset column
            values = data[column].dropna()  # Remove missing values
            
            labelStr = f'mu: {values.mean():.3f}\nmedian: {values.median():.3f}\nstd: {values.std():.3f}\nmin: {values.min():.3f}\nmax: {values.max():.3f}'
            
            # Set the label and add the legend
            subplot.set_ylabel("Frequency")
            subplot.legend([labelStr])

plt.tight_layout()
plt.show()
```

```python
print(f"Minimum RestingBP: {data['RestingBP'].min()}")
print(f"Minimum Cholesterol: {data['Cholesterol'].min()}")
print(f"Count of RestingBP = 0: {len(data[data['RestingBP'] == 0])}")
print(f"Count of Cholesterol = 0: {len(data[data['Cholesterol'] == 0])}")
# Replace RestingBP = 0 with the median
data.loc[data['RestingBP'] == 0, 'RestingBP'] = data['RestingBP'].median()
data.loc[data['Cholesterol'] == 0, 'Cholesterol'] = data['Cholesterol'].mean() #If cholesterol values are normally distributed, using the mean is also a valid option: 


ax = data.hist(bins=25, figsize=(15, 15))

# Iterate over each row of subplots
for row in ax:
    for subplot in row:
        # Get the column name from the subplot's title
        column = subplot.get_title()
        
        if column:  # Check if the subplot has a title
            # Calculate statistics directly from the dataset column
            values = data[column].dropna()  # Remove missing values
            
            labelStr = f'mu: {values.mean():.3f}\nmedian: {values.median():.3f}\nstd: {values.std():.3f}\nmin: {values.min():.3f}\nmax: {values.max():.3f}'
            
            # Set the label and add the legend
            subplot.set_ylabel("Frequency")
            subplot.legend([labelStr])

plt.tight_layout()
plt.show()
```

```python
# Replace 'Sex' column with cleaned version
data['Sex'] = data_cleaned_Sex

# One-hot encode ChestPainType and convert booleans to ints
chest_pain_dummies = pd.get_dummies(data['ChestPainType'], prefix='ChestPain').astype(int)

# Add one-hot columns and drop the original
data = pd.concat([data, chest_pain_dummies], axis=1)
data = data.drop(columns=['ChestPainType'])

# Assign cleaned categorical columns
data['RestingECG'] = data_cleaned_RestingECG
data['ExerciseAngina'] = data_cleaned_ExerciseAngina
data['ST_Slope'] = data_cleaned_ST_Slope

# Compute correlation matrix
my_matrix = data.select_dtypes(include='number').corr(method="spearman").round(2)
my_matrix.style.background_gradient(cmap="Greens")
```
